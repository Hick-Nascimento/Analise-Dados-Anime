```python
# Distribuição de popularidade com linha de tendência
plt.figure(figsize=(10, 6))
sns.histplot(base['Popularity'], bins=30, kde=True, color='red')
sns.kdeplot(base['Popularity'], color='blue')  # Adiciona uma linha de tendência
plt.title('Popularity Distribution with Trend Line')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Definição dos dados
top_producers = base['Producers'].str.split(',').explode().value_counts().head(10)

# Gráfico com paleta de cores diferente
plt.figure(figsize=(10, 6))
sns.barplot(x=top_producers.values, y=top_producers.index, palette='tab10')  # Usando a paleta de cores 'tab10'
plt.title('Top 10 Produtores por contagem')
plt.xlabel('Count')
plt.ylabel('Produtor')
plt.show()
```
